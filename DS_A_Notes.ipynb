```
#stack implementation
#follows LIFO - Last in first out.
class Stack:
  def __init__(self):
    self.stack = []
  def isEmpty(self):
    return self.stack == []
  def push(self, data):
    self.stack.append(data)
  def pop(self):
    data = self.stack[-1]
    del self.stack[-1] #can also be self.stack.pop() as this function is inbuilt
    return data
  def peek(self):
    return self.stack[-1]
  def sizeStack(self):
    return len(self.stack)
```

```
#queue implementation
#follows FIFO - first in first out 
class Queue:
  def __init__(self):
    self.queue = []
  def isEmpty(self):
    return self.queue == []
  def enqueue(self, data):
    self.queue.append(data)
  def dequeue(self):
    data = self.queue[0]
    del self.queue[0] #can also use self.queue.pop(0)
    return data 
  def sizeQueue(self):
    return len(self.queue)
```

```
#binary search tree



class Node(object):
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None

class BinarySearchTree(object):
    def __init__(self):
        self.root = None
  
    def insert(self, data):
        if not self.root:
            self.root = Node(data)
        else:
            self.insertNode(data, self.root)
    
    '''
    recursive insert function. starts at root and checks if a child exists at
    either the right or left of the parent. If there is no node, existing at a position
    which the data can fill, then a new node object is instantiated and added to the 
    BST
    '''
    def insertNode(self, data, node):
        if data > node.data:
            if not node.rightChild:
                node.rightChild = Node(data)
            else:
                self.insertNode(data, node.rightChild)
        else:
            if not node.leftChild:
                node.leftChild = Node(data)
            else:
                self.insertNode(data, node.leftChild)

    #recursive remove, sets the root into a modified version of the original tree
    def remove(self, data):
        if self.root:
            self.root = self.removeNode(data, self.root)
    def removeNode(self, data, node):
        if not node:
            return node

        #traversing the tree recursively to find the node that needs to be deleted
        #Once the node found, the modified node will be returned after all transformations are applied to it
        #this causes the initial node to update as well, which basically propogates backwards until the root
        if data < node.data:
            node.leftChild = self.removeNode(data, node.leftChild)
        elif data > node.data:
            node.rightChild = self.removeNode(data, node.rightChild)
        else:

            #no right or left child
            if not node.leftChild and not node.rightChild:
                print("removing leaf node")
                del node
                return None
            #right child. Results in the parent node's right being reassigned to the right child of the deleted node 
            elif not node.leftChild: 
                print("removing node with a single right child")
                tempNode = node.rightChild
                del node
                return tempNode
            #left child. Results in the parent node's left being reassigned to the left child of the deleted node 
            elif not node.rightChild:
                print("removing a node with a single left child")
                tempNode = self.leftChild
                del node
                return tempNode
            
            #case where the node has two children
            print("removing the node with two children...")

            #replace the value of the desired node for deletion with the predecessor
            tempNode = self.get_predecessor(node.leftChild)
            node.data = tempNode.data

            #remove the predecessor from the left subtree recursively 
            #and reassign the left child of the deleted node (now holding the value of the predecessor) to this modified left subtree
            node.leftChild = self.removeNode(tempNode.data, node.leftchild)

        #this is really important as it passes the modified node back to the parent node recursively
        return node
    def get_predecessor(self, node):
        #keep travelling right to get the rightmost node in the left subtree
        if node.rightChild:
            return self.getPredecessor(node.rightChild)
        return node

    #get min value, leftmost node in tree
    def getMinValue(self):
        if self.root:
            return self.getMin(self.root)
    def getMin(self, node):
        if node.leftChild:
            return self.getMin(node.leftChild)
        return node.data

    #get max value, rightmost node in tree
    def getMaxValue(self):
        if self.root:
            return self.getMax(self.root)
    def getMax(self, node):
        if node.rightChild:
            return self.getMax(node.rightChild)
        return node.data

    def traverse(self):
        if self.root:
            self.traverseInOrder(self.root)

    #inorder traversal, first print leftmost child, then middle, then right. If a subtree exists then this function will handle it accordingly through recursion 
    def traverseInOrder(self, node):
        if node.leftChild:
            self.traverseInOrder(node.leftChild)

        #this will execute regardless of the amount of children a node has, meaning if a middle node is passed to this function, a middle node will be printed 
        print(str(node.data))

        if node.rightChild:
            self.traverseInOrder(node.rightChild)
```
